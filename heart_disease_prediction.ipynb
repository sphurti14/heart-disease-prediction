```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
df['target'] = df['num'].apply(lambda x: 0 if x == 0 else 1)
df[['num', 'target']].head()
```

```python
df = df.drop(['id', 'dataset', 'num'], axis=1)
df.head()
```

```python
categorical_val = []
continuous_val = []

for column in df.columns:
    if df[column].nunique() <= 10 and column != 'target':
        categorical_val.append(column)
    elif column != 'target':
        continuous_val.append(column)

categorical_val, continuous_val
```

```python
dataset = pd.get_dummies(df, columns=categorical_val)
dataset.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

dataset[continuous_val] = scaler.fit_transform(dataset[continuous_val])
dataset.head()

```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
dataset.iloc[:, :] = imputer.fit_transform(dataset)

# Quick check
dataset.isna().sum().sum()
```

```python
X = dataset.drop('target', axis=1)
y = dataset['target']

X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

accuracy
```
